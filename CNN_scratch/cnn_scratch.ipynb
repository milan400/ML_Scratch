```python
import skimage.data
from PIL import Image
from matplotlib import pyplot as plt
import numpy as np
from scipy import ndimage
```

Reading the image

```python
img = skimage.data.chelsea()
print(img.shape)
```

```python
plt.imshow(img, interpolation='nearest')
plt.show()
```

Converting the image into gray

```python
img = skimage.color.rgb2gray(img)
print(img.shape)
```

```python
plt.imshow(img,interpolation='nearest')
plt.show()
```

```python
img.shape
```

```python
def apply_kernel(img,kernel):
    return(ndimage.convolve(img,kernel,mode="constant",cval=0.0))
```

```python
kernel = np.array([[1,0,-1],[2,0,-2],[1,0,-1]])
img = apply_kernel(img,kernel)
```

```python
print(img.shape)
```

```python
plt.imshow(img,interpolation='nearest')
plt.show()
```

```python
pool_image = skimage.measure.block_reduce(img,(2,2),np.max)
print(pool_image.shape)
```

```python
plt.imshow(pool_image,interpolation='nearest')
plt.show()
```

```python
flattened_image = pool_image.reshape(-1,1)
```
