```python
from sklearn.datasets import load_iris
```

```python
iris=load_iris()
```

```python
iris
```

```python
import pandas as pd
```

```python
dataset=pd.DataFrame(data=iris['data'],columns=['sepal length','sepal width','petal length','petal width'])
```

```python
dataset
```

```python
type(dataset)
```

```python
dataset['label']=iris['target']
dataset.head()
```

```python
X=dataset.drop('label',axis=1)
```

```python
X
```

```python
Y=dataset['label']
```

```python
Y
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler= StandardScaler()
```

```python
scaler.fit(X)
```

```python
scaled_features= scaler.transform(X)
```

```python
scaled_features
```

```python
df_feat=pd.DataFrame(scaled_features,columns=dataset.columns[:-1])
```

```python
df_feat
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test= train_test_split(scaled_features,dataset['label'],test_size=0.30)
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
knn= KNeighborsClassifier(n_neighbors=2)
```

```python
knn.fit(X_train,y_train)
```

```python
pred = knn.predict(X_test)
```

```python
pred
```

```python
from sklearn.metrics import classification_report,confusion_matrix
```

```python
print(classification_report(y_test,pred))
```
